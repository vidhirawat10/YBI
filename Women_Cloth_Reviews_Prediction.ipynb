**WOMEN CLOTH REVIEWS PREDICTION WITH MULTI NOMIAL NAIVE BAYES**

**OBJECTIVE** - The objective of the project is to develop a machine learning model that can accurately predict the sentiment of reviews (positive or negative) for women's clothing. By analyzing the text of these reviews using a Multinomial Naive Bayes algorithm, the model aims to classify new reviews based on their sentiment, helping businesses understand customer opinions and improve their products and services.

**DATA SOURCE** - YBI Foundation Github's dataset

**1. Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**2. Importing Dataset**

```python
df = pd.read_csv("https://raw.githubusercontent.com/YBIFoundation/ProjectHub-MachineLearning/main/Women%20Clothing%20E-Commerce%20Review.csv")
```

**3. Describing Data - Data Visualization**

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

**4. Data Preprocessing**

```python
df.isna().sum()  #checking if any null value present in dataset or not
```

```python
df[df['Review']==""] = np.NaN     #Filling missing values in the Review column with the value No review is given.
df['Review'].fillna("No review is given", inplace=True)
df.isna().sum()
```

```python
df['Review']
```

**5. Defining Target Variable (y) and Feature Variables (X)**

```python
df.columns
```

```python
X = df['Review']
y = df['Review']
df['Rating'].value_counts()
```

**6. Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=  0.7, random_state = 2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

**7. Modeling**

Getting Tokens from Feature Text

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(lowercase = True, analyzer='word', ngram_range=(2, 3), stop_words = 'english', max_features = 5000)
```

```python
X_train = cv.fit_transform(X_train)
cv.get_feature_names_out()
```

```python
X_train.toarray()
```

```python
X_test = cv.fit_transform(X_test)
```

```python
cv.get_feature_names_out()
```

```python
X_test.toarray()
```

**8. Model Train**

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train, y_train)
```

**Model Prediction**

```python
y_pred = model.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

Probaility of Each Predicted Class

```python
model.predict_proba(X_test)
```

**Model Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

Getting Tokens from Feature Text

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(lowercase=True, analyzer='word', ngram_range=(2, 3), stop_words='english', max_features=50000)
X_train = cv.fit_transform(X_train)
X_test = cv.fit_transform(X_test)
```

Model re-training

**8. Model prediction**

```python
y_pred = model.predict(X_test)
y_pred.shape
```

```python
y_pred
```

**9. Model evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
